```python
def evl(board, player):
    # this function will calculate the score on the board, if a move is performed
    # give score for each of piece and calculate the score for the chess board
    score = 0
    # go through board
    for r in range(0, 8):
        for c in range(0, 8):
            piece = board[r][c]
            #if empty
            if(piece == '.'):
                #continue loop
                continue
            # else if : for piece of player
            if((player == 'black' and piece.islower()) or (player == 'white' and piece.isupper())):
                # add score
                pieceScore = GetPieceScore(piece)
                score = score + pieceScore
            # else : for piece of enemy
            else:
                # minus score
                pieceScore = GetPieceScore(piece)
                score = score - pieceScore

    #return score
    return score

def GetMinMaxMove2(board, player):
    # return the best move for the current player using the MinMax strategy
    # to get the allocated points, searching should be 2-ply (one Max and one Min)

    # Following is the setup for a 2-ply game

    #init best Move candidates
    bestMoves = []
    #init maxScore
    maxScore = -99999
    # get all movable pieces
    movablePieceSquares = GetPiecesWithLegalMoves(board, player)
    # for each pieceSquare in movablePieceSquares
    for pieceSquare in movablePieceSquares:
        
        # get all legal moves for given piece
        moves = GetListOfLegalMoves(board, player, pieceSquare)
        # for each move in moves
        for move in moves:
            # perform the move temporarily
            oneStepBoard = MovePiece(board, move)
            
            #init best EnemyMove
            bestEnemyMove = ()
            #init minScore
            minScore = 999999

            # get enemyPlayer label
            enemyPlayer = 'black' if player == 'white' else 'white'
            # get enemy movable piece for after player move
            enemyMovablePieceSquares = GetPiecesWithLegalMoves(oneStepBoard, enemyPlayer)
            # for each enemyPieceSquare in enemyMovablePieceSquares
            for enemyPieceSquare in enemyMovablePieceSquares:
                # get all legal moves for given piece
                enemyMoves = GetListOfLegalMoves(oneStepBoard, enemyPlayer, enemyPieceSquare)
                # for each enemyMove in enemyMoves
                for enemyMove in enemyMoves:
                    # perform the enemyMove temporarily
                    twoStepBoard = MovePiece(oneStepBoard, enemyMove)
                    # get score after enemy move
                    res = evl(twoStepBoard, player)
                    # update the bestEnemyMove -- this is the MIN player trying to minimize from the 'res' evaluation values
                    if(minScore > res):
                        bestEnemyMove = enemyMove
                        minScore = res
            # update the bestMove -- this is the MAX player trying to maximize from the 'bestEnemyMove' evaluation values
            if(maxScore < minScore):
                bestMoves = [move]
                maxScore = minScore
            elif(maxScore == minScore):
                # add move to bestMoves
                bestMoves.append(move)
    
    #random pick bestMove from bestMove candiations
    bestMove = random.choice(bestMoves)

    # OPTIONAL -- sometimes automated chess keeps on performing the moves again and again
    # e.g., move king left one square and then move king back - repeat
    # For this you will need to remember the previous move and see if the current best move is not the same and opposite as the previous move
    # If so, pick the second best move instead of the best move

    return bestMove






def GetMinMaxMove(board, player, depth = 4):
    # return the best move and second best move for the current player using the MinMax strategy
    # to get the allocated points
    # searching should be 4-ply (one Max and one Min one Max one Min) if depth = 4
    # 2-ply(one Max and one Min) if depth = 2
     
    #init best Move candidates and secondBestMoves
    bestMoves = []
    secondBestMoves = []

    #init maxScore
    maxScore = -99999
    secondMaxScore = -99999
    #init alpha, beta
    alpha = -99999
    beta = 99999

    # get all movable pieces
    movablePieceSquares = GetPiecesWithLegalMoves(board, player)
    # for each pieceSquare in movablePieceSquares
    for pieceSquare in movablePieceSquares:
        
        # get all legal moves for given piece
        moves = GetListOfLegalMoves(board, player, pieceSquare)
        # for each move in moves
        for move in moves:
            # perform the move temporarily
            newBoard = MovePiece(board, move)
            # get enemy label
            enemy = 'black' if player == 'white' else 'white'
            # get enemy minmax
            enemyMinmax = getMinmaxValue(depth - 1, newBoard, enemy, alpha, beta, False)
            #enemyMinmax > maxScore
            if(enemyMinmax > maxScore):
                # update bestMoves and secondBestMoves
                secondBestMoves = bestMoves
                bestMoves = [move]
                # update maxScore and secondMaxScore
                secondMaxScore = maxScore
                maxScore = enemyMinmax
            # else if enemyMinmax == maxScore
            elif (enemyMinmax == maxScore):
                # add move to bestMoves
                bestMoves.append(move)
            # else if enemyMinMax > secondMaxScore
            elif( enemyMinmax > secondMaxScore):
                #update secondBestMoves and secondMaxScore
                secondBestMoves = [move]
                secondMaxScore = enemyMinmax
            # else if enemyMinMax = secondMaxScore
            elif(enemyMinmax == secondMaxScore):
                # add move to secondBestMoves
                secondBestMoves.append(move)
    
    #pick bestMove and secondBestMove
    # if there are more than one bestMoves
    if(len(bestMoves) > 1):
        #return random 2 bestMoves
        return random.choices(bestMoves, k = 2)
    else:
        #random pick secondBestMove
        secondBestMove = random.choice(secondBestMoves)
        return [bestMoves[0], secondBestMove]


# get minmax value of the given depth and board
# return maximum value if is_maximizing = true else return minimum
def getMinmaxValue(depth, board, player, alpha, beta, is_maximizing):
    # if depth = 0
    if(depth == 0):
        #return the score of board
        return evl(board, player)

    #init result
    res = -99999 if is_maximizing else 99999

    # get all movable pieces
    movablePieceSquares = GetPiecesWithLegalMoves(board, player)
    for pieceSquare in movablePieceSquares:
        # get all legal moves for given piece
        moves = GetListOfLegalMoves(board, player, pieceSquare)
        # for each move in moves
        for move in moves:
            # perform the move temporarily
            newBoard = MovePiece(board, move)
            #get Enemy label
            enemy = 'black' if player == 'white' else 'white'
            #get Enemy minmax
            enemyMinMax = getMinmaxValue(depth - 1, newBoard, enemy, alpha, beta, not is_maximizing)
            # update max if is_maximizing = true else update min
            res = max(enemyMinMax, res) if is_maximizing else min(enemyMinMax, res)
            # update result, alpha, beta
            # if is_maximizing = true
            if(is_maximizing):
                res = max(enemyMinMax, res)
                alpha = max(alpha, enemyMinMax)
                #if beta <= alpha
                if(beta <= alpha):
                    #stop searching
                    return res
            else:
                res = min(enemyMinMax, res)
                beta = min(enemyMinMax, beta)
                if(beta <= alpha):
                    return res
    return res


# Get score for given piece
def GetPieceScore(piece):
    # for pawn give 10 score
    if piece.lower() == 'p':
        return 10
    # for knight give 30 score
    if piece.lower() == 't':
        return 30
    # for bishop give 30 score
    if piece.lower() == 'b':
        return 30
    # for rook give 50 score
    if piece.lower() == 'r':
        return 50
    # for Queen give 90 score
    if piece.lower() == 'p':
        return 90
    # for King give 900 score
    if piece.lower() == 'p':
        return 900
    
    #else
    return 0
```
